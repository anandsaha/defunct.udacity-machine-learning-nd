### Student Admission

Given student's gpa, grep score and rank, predict if the student will get admission or not

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras 
from sklearn import model_selection

%matplotlib inline
```

```python
data = pd.read_csv('student_data.csv')
data.head()
```

### Split between X and y

```python
X = data.loc[:, ['gre', 'gpa', 'rank']]
y = data.loc[:, ['admit']]
```

### Process X

```python
# Fill all NA's with 0's
X = X.fillna(0)

# Normalize gre and gpa values
X['gpa'] = X['gpa'] / 4
X['gre'] = X['gre'] / 800

# One hot encode rank
X = pd.get_dummies(X, columns=['rank'])
X = X.astype('float32')
X = X.values
```

### Process y

```python
y = keras.utils.to_categorical(y["admit"],2)
```

### Generate training and testing sets

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, random_state=42, test_size=0.2)
```

### Create MLP

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout

model = Sequential()
# Layer 1
model.add(Dense(32, input_shape=(7,)))
model.add(Activation('relu'))
model.add(Dropout(.2))
# Layer 2
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(.2))
# Layer 3
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(.2))
# Output
model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=16, epochs=500, validation_split=0.3)
```

```python
model.evaluate(X_test, y_test)
```
